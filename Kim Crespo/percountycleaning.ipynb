```python
# import dependencies
import pandas as pd
from pathlib import Path



# paths for .csv file
northern_county_market = Path("../Resources/northern_county_market.csv")
central_county_market = Path("../Resources/central_county_market.csv")
southern_county_market = Path("../Resources/southern_county_market.csv")

allcounties = pd.read_csv("../Resources/fullcleaned_county_market_tracker.csv")
allcounties.head()
```

```python
print(allcounties['County'].unique())
```

```python
# filter by northern counties
northern_counties = ['Sussex County', 'Passaic County', 'Bergen County', 'Warren County',
                     'Morris County', 'Essex County', 'Hudson County']

north_counties_df = allcounties[allcounties['County'].isin(northern_counties)]

# adding region column for 'North'
north_region = 'North'
north_counties_df.insert(north_counties_df.columns.get_loc('County') + 1, 'Region', north_region)

north_counties_df.head()
# create new csv (northern)
north_counties_df.to_csv(northern_county_market, index=False)
```

```python
# filter by central counties
central_counties = ['Hunterdon County', 'Somerset County', 'Union County', 'Middlesex County',
                    'Mercer County', 'Monmouth County']
            
central_counties_df = allcounties[allcounties['County'].isin(central_counties)]

# adding region column for 'Central'
central_region = 'Central'
central_counties_df.insert(central_counties_df.columns.get_loc('County') + 1, 'Region', central_region)

central_counties_df.head()

#create new csv (central)
central_counties_df.to_csv(central_county_market, index=False)
```

```python
# filter by southern counties
southern_counties = ['Ocean County', 'Burlington County', 'Camden County',
                     'Gloucester County', 'Salem County', 'Cumberland County',
                     'Cape May County', 'Atlantic County']

southern_counties_df = allcounties[allcounties['County'].isin(southern_counties)]

# adding region column for 'Southern'
southern_region = 'Southern'
southern_counties_df.insert(southern_counties_df.columns.get_loc('County') + 1, 'Region', southern_region)
southern_counties_df.head()

# create new csv (southern)
southern_counties_df.to_csv(southern_county_market, index=False)
```
