```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df =  pd.read_csv("../Resources/fullcleaned_county_market_tracker_no_symbols.csv")
df
```

```python
five_counties = ['Essex County', 'Passaic County', 'Somerset County', 'Salem County', 'Sussex County']

# Datetime
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])

# Filter
df_filtered = df[df['County'].isin(five_counties)]
df_filtered
```

```python
df_2022 = df_filtered[df_filtered['Start Date'] == '2022-01-01']
df_2023 = df_filtered[df_filtered['Start Date'] == '2023-01-01']
df_2024 = df_filtered[df_filtered['Start Date'] == '2024-01-01']
```

```python
# Rename columns
df_2022 = df_2022[['County', 'Property Type', 'Median Price per Sqft']].rename(columns={'Median Price per Sqft': 'Median Price per Sqft_2022'})
df_2023 = df_2023[['County', 'Property Type', 'Median Price per Sqft']].rename(columns={'Median Price per Sqft': 'Median Price per Sqft_2023'})
df_2024 = df_2024[['County', 'Property Type', 'Median Price per Sqft']].rename(columns={'Median Price per Sqft': 'Median Price per Sqft_2024'})
```

```python
# Merge
df_merged_2022_2023 = pd.merge(df_2022, df_2023, on=['County', 'Property Type'])
df_merged_2023_2024 = pd.merge(df_2023, df_2024, on=['County', 'Property Type'])
```

```python
# Percent Change
df_merged_2022_2023['Percent Change 2022-2023'] = ((df_merged_2022_2023['Median Price per Sqft_2023'] - df_merged_2022_2023['Median Price per Sqft_2022']) / df_merged_2022_2023['Median Price per Sqft_2022']) * 100
df_merged_2023_2024['Percent Change 2023-2024'] = ((df_merged_2023_2024['Median Price per Sqft_2024'] - df_merged_2023_2024['Median Price per Sqft_2023']) / df_merged_2023_2024['Median Price per Sqft_2023']) * 100
```

```python
# Pivot tables for separate plots
pivot_table_2022_2023 = df_merged_2022_2023.pivot_table(values='Percent Change 2022-2023', index='County', columns='Property Type', fill_value=0)
pivot_table_2023_2024 = df_merged_2023_2024.pivot_table(values='Percent Change 2023-2024', index='County', columns='Property Type', fill_value=0)

# Plot for 2022 to 2023
fig, ax = plt.subplots(figsize=(20, 10))
pivot_table_2022_2023.plot(kind='bar', ax=ax)
ax.set_title('Percent Change of Median Price per Sqft by Property Type and County (2022 to 2023)')
ax.set_xlabel('County')
ax.set_ylabel('Percent Change (%)')
ax.legend(title='Property Type', loc='upper left')
plt.tight_layout()
plt.show()

# Plot for 2023 to 2024
fig, ax = plt.subplots(figsize=(20, 10))
pivot_table_2023_2024.plot(kind='bar', ax=ax)
ax.set_title('Percent Change of Median Price per Sqft by Property Type and County (2023 to 2024)')
ax.set_xlabel('County')
ax.set_ylabel('Percent Change (%)')
ax.legend(title='Property Type', loc='upper left')
plt.tight_layout()
plt.show()
```
